**Apply the kNN classifier algorithm. Apply K=5 and predict the flower of the** **given datapoint.**

**Predict the test datapoint ={'sepal length' : [4.6], 'sepal_width' : [3.4],** **'petal_length' : [1.4], 'petal_width' : [0.3], }**

**Actual output: 'flower' : ‘setosa’ Finding the predicted output for above test**  **datapoint**

```python
# Load Iris Data
iris = load_iris()
```

```python
iris_dataframe = pd.DataFrame(data= iris.data, columns= iris.feature_names)
print(iris_dataframe)
```

```python
iris.target
```

```python
iris.target_names
```

```python
target_df = pd.DataFrame(data= iris.target, columns= ['species'])
```

```python
target_df = pd.DataFrame(data= iris.target, columns= ['species'])
def converter(specie):
    if specie == 0:
        return 'setosa'
    elif specie == 1:
        return 'versicolor'
    else:
        return 'virginica'
target_df['species'] = target_df['species'].apply(converter)
target_df
```

```python
iris_dataframe = pd.concat([iris_dataframe, target_df], axis= 1)
iris_dataframe
```

```python
iris_dataframe.describe()
```

```python
target_df = pd.DataFrame(columns= ['species'], data= iris.target)
target_df
```

```python
X = iris_dataframe[iris.feature_names].values
y = iris_dataframe['species'].values
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of our model is ' + str(round(accuracy, 2)) + ' %.')
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
cr = classification_report(y_test, y_pred)
print("Classification Report:",)
print (cr)
```

```python
new_datapoint = {'sepal_length' : [4.6], 'sepal_width' : [3.4], 'petal_length' : [1.4], 'petal_width' : [0.3] }
test_datapoint=pd.DataFrame(data=new_datapoint)
predict_testdatapoint = classifier.predict(test_datapoint)
predict = predict_testdatapoint[0]
print('The predicted flower for the given data point from iris dataset is',iris.target_names[predict])
```

We have obtained the accuracy from the KNN classifier model. The obtained accuracy is 96.67 %. It is good accuracy. So we can say that the model is fitted correctly. The predicted output is the setosa and the actual output is also setosa. That means the model is correctly predicting the output of the given test data point.
